### SORTEIO DA FRUTA QUE NAO IRA COMER NO DIA DE SAO JOSE

Um dia, uma podcaster chamada Camila Frender contou no Instagram dela sobre uma tradição curiosa do Dia de São José: sortear uma fruta e ficar um ano inteiro sem comê-la. Dizem que, ao fazer isso e deixar um pedido, ele pode se realizar.

Resolvi testar. No primeiro ano saiu melancia. No segundo, limão — esse foi sofrido, tive até de inventar que tinha alergia para escapar de algumas comidas. Agora, no próximo 19 de março, vou fazer novamente. Curiosamente, todos os anos o pedido acabou se realizando.

Como estudante de Ciência da Computação, claro que não podia simplesmente sortear de qualquer forma… acabei criando o meu próprio código para fazer o sorteio.

O dia 19 de março celebra São José, tradicionalmente visto como símbolo de proteção da família e do trabalho. A data também coincide com a chegada da primavera no hemisfério norte, época associada a renovação e novos ciclos — talvez por isso faça tanto sentido ligar o dia a pequenos rituais de esperança e pedidos para o ano que começa.

```python
#Importar o Random
import random

#Colocar a lista de frutas que come
frutas = ["Abacate", "Abacaxi",  "Azeitona", "Açaí", "Banana", "Cacau", "Coco", "Guaraná", "Laranja", "Limão", "Maçã", "Mamão", "Manga", "Maracujá", "Morango", "Melancia", "Melão",
"Pêra", "Pêssego", "Tangerina", "Tomate", "Uva"]

#Imprimir de forma aleatoria a fruta que nao ira comer depois a partir do dia de Sao Jose.

print("Fruta que nao ira comer no dia de Sao Jose: " + random.choice(frutas))
```
